```
import os
import numpy as np
import pandas as pd
```

```
# Generate pandas dataframe.
DIR = '/gdrive/My Drive/data_all/'
root = os.listdir('/gdrive/My Drive/data_all')

books_list = list()
for genre in root:
  for label in os.listdir(DIR + genre):
    for title in os.listdir(DIR + genre + '/' + label):
      with open(DIR + genre + '/' + label + '/' + title, 'rb') as f:
        data = f.read()
      books_list.append({'title': title, 'genre': genre, 'label': label, 'content': data.decode('utf-8', 'ignore')})

books_df = pd.DataFrame(books_list)
```

```
# books_df.to_csv('books.csv')
books_df.head()
```

```
# Clean content
def remove_punct_lower(text):
    PUNCT = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
    no_punct = ""
    for char in text:
        char = str(char)
        if char not in PUNCT:
            no_punct = no_punct + char
    return ' '.join(no_punct.split()).lower() 
```

```
books_df['content'] = books_df['content'].apply(remove_punct_lower)
books_df.head()
```

```
with open('/gdrive/My Drive/Datasets/ef.csv', 'w') as f:
    books_df.to_csv(f)
```
